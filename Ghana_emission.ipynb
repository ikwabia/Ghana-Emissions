```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
#checking data from 2000 till date
Year_20 = df[df['Year'] >= 2000]
```

```python
Year_20
```

```python
#selecting only data for Ghana from 2000
Ghana_20 = Year_20[Year_20['Entity'] == 'Ghana']
```

```python
Ghana_20
```

```python
# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(Ghana_20['Year'], Ghana_20['Carbon monoxide (CO) emissions'], marker='o', linestyle='-', color='b', label='CO Emissions')

# Customizing the plot
plt.title('Carbon Monoxide (CO) Emissions Over Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO Emissions', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(True)

# Removing the top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Display the plot
plt.show()
#plt.savefig('CO_Emissions.png')
```

```python
import matplotlib.pyplot as plt

# Plotting the stacked area chart
plt.figure(figsize=(10, 5))
plt.stackplot(Ghana_20['Year'], Ghana_20['Carbon monoxide (CO) emissions'], Ghana_20['Nitrogen oxide (NOx)'],Ghana_20['Sulphur dioxide (SO₂) emissions'],Ghana_20['Non-methane volatile organic compounds (NMVOC) emissions'], labels=['CO Emissions', 'NOx Emissions','SO₂ emissions','NMVOC emissions'], colors=['r','y', 'g','b'], alpha=0.5)

# Customizing the plot
plt.title(' EMISSION OVER THE YEARS', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Emissions', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(loc='upper left')
plt.grid(True)

# Removing the top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Display the plot
plt.show()
#plt.savefig('Emissions_over_years.png')

```

```python
# total emissions yearly
ranked_emissions = Ghana_20.sort_values(by='Total Emissions', ascending=False)

ranked_emissions['Rank'] = range(1, len(ranked_emissions) + 1)
# Print the ranked total emissions
print(ranked_emissions[['Year', 'Total Emissions', 'Rank']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 8))
barplot = sns.barplot(x='Year', y='Total Emissions', data=ranked_emissions, palette='viridis')

# Set the title and labels
plt.title('Total Emissions Ranking')
plt.xlabel('Rank')
plt.ylabel('Total Emissions')
plt.xticks(rotation=45)

# Show the plot
plt.show()
#plt.savefig('Year_emission.png')

```

```python
T_Co = Ghana_20['Carbon monoxide (CO) emissions'].sum()
T_No = Ghana_20['Nitrogen oxide (NOx)'].sum()
T_So = Ghana_20['Sulphur dioxide (SO₂) emissions'].sum()
T_Oc = Ghana_20['Organic carbon (OC) emissions'].sum()
T_NM = Ghana_20['Non-methane volatile organic compounds (NMVOC) emissions'].sum()
T_Bc = Ghana_20['Black carbon (BC) emissions'].sum()
T_NH = Ghana_20['Ammonia (NH₃) emissions'].sum()


# Create a new DataFrame for plotting
total_emissions_df = pd.DataFrame({
    'Emission Type': ['CO', 'NOx', 'SO','OC','NMVOC','BC','NH'],
    'Total Emissions': [T_Co,T_No,T_So,T_Oc,T_NM,T_Bc,T_NM]
})
```

```python
plt.figure(figsize=(10, 6))
plt.bar(total_emissions_df['Emission Type'], total_emissions_df['Total Emissions'], color=['red', 'green', 'blue','yellow','violet','black','indigo'])

plt.xticks(rotation=45)

# Set the title and labels
plt.title('Total Emissions by Type from 2000 to 2022')
plt.xlabel('Emission Type')
plt.ylabel('Total Emissions')

# Show the plot
plt.tight_layout() # Adjust the layout to fit the rotated labels
plt.show()
#plt.savefig('total_emissions_by_type.png')
#files.download('total_emissions_by_type.png')
```
